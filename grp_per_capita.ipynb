```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pandas as pd
import matplotlib.colors as mcolors

# 假设 map_data 和 sub_ratio_df 已经定义，且数据格式正确
# 你的 _plot 函数：
def _plot(ax, year: int):
    import matplotlib
    matplotlib.rcParams['font.size'] = 10  # 为子图调整字体大小
    matplotlib.rcParams['font.weight'] = 'light'
    df = map_data.copy()
    ratios = []
    for r in df['ADM1_ZH']:
        try:
            ratio = sub_ratio_df[sub_ratio_df['region'] == r][str(year)].values[0]
            ratios.append(ratio)
        except:
            ratios.append(0)
    df['ratios'] = pd.DataFrame(ratios)
    
    # 绘图到指定的轴 ax
    ax.set_axis_off()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("bottom", size="4%", pad=0.1)  # 调整间距和大小
    df.plot(column="ratios", cmap='Greens', legend=True, 
            norm=mcolors.Normalize(vmin=0, vmax=200000), ax=ax, cax=cax,
            legend_kwds={"label": "GRP per capita (CNY)", 
                         "orientation": "horizontal"})

# 绘制2012到2023年所有子图
years = range(2012, 2024)
n_cols = 4  # 每行子图数
n_rows = -(-len(years) // n_cols)  # 计算行数（向上取整）
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows * 4))

# 平铺年份的子图
axes = axes.flatten()  # 转为1D数组，方便索引
for i, year in enumerate(years):
    _plot(axes[i], year)
    axes[i].set_title(f"{year}", fontsize=14)

# 隐藏多余的子图
for j in range(len(years), len(axes)):
    axes[j].set_visible(False)

plt.tight_layout()
plt.show()
```
