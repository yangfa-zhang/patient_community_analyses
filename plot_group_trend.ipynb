```python
#颜色调整；纵坐标用： “Fold change in the normalized number of users” ; 西藏fold change数据可以不用; 左右两幅图 标记a b

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Replace 'your_file.csv' with the path to your .csv file
file_path = 'normalized_posts_with_provinces1.csv'

# Read the .csv file into a DataFrame
df = pd.read_csv(file_path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 假设 foldchange_values 已经是一个pandas Series
foldchange_values = df['Foldchange']
group_labels = ['Group 1'] * 3 + ['Group 2'] * (len(foldchange_values) - 3)

# Create a new DataFrame for the plot
plot_data = pd.DataFrame({
    'Foldchange': foldchange_values,
    'Group': group_labels
})

# Step 3: Draw the violin plot with custom colors
plt.figure(figsize=(12, 8))
# sns.violinplot(x='Group', y='Foldchange', data=plot_data, palette={'Group 1': '#E89DA0', 'Group 2': '#88CEE6'})
sns.violinplot(x='Group', y='Foldchange', data=plot_data, palette={'Group 1': '#007BFF', 'Group 2': '#FF6F00'})

plt.xticks(rotation=0, fontsize=22)
plt.yticks(fontsize=22)
plt.xlabel('',fontsize=20)
plt.ylabel('Fold change in user participation rate',fontsize=24)
plt.show()
```
